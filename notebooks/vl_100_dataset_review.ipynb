````python
scenes = set()
for file in list(glob.glob(f"{JSON_PATH}/*/*.json")):                                       
  with open(file) as f:
    json_example = json.load(f)
    scenes.add(json_example['info']['scenes'])
  
````

````python
print(scenes) #wtf?????
````

````python
dimensions = set()
for file in list(glob.glob(f"{JSON_PATH}/*/*.json")):                                       
  with open(file) as f:
    json_example = json.load(f)
    dimensions.add((json_example['info']['height'],json_example['info']['width']))
print(dimensions)
````

````python
num_lines_in_frame = set()
for file in list(glob.glob(f"{JSON_PATH}/*/*.json")):                                       
  with open(file) as f:
    json_example = json.load(f)
    num_lines_in_frame.add(len(json_example['annotations']['lane']))
print(num_lines_in_frame)
````

````python
max_length_of_lines = set()
for file in list(glob.glob(f"{JSON_PATH}/*/*.json")):                                       
  with open(file) as f:
    json_example = json.load(f)
    max_length_of_lines.add(max([len(x["points"]) for x in json_example['annotations']['lane']] or [0]))
print(max_length_of_lines)
````
